```python
import pandas as pd
expedu = pd.read_csv('./expendonedu.csv')# expenditure on education
enrollment = pd.read_csv('./schoolenrollment.csv')# school enrollment
literacy = pd.read_csv('./literacy.csv')# literacy rates
meanschool = pd.read_csv('./meanschool.csv')# mean years of schooling
```

```python
countries = ['Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Czechia',
             'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece',
             'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan',
             'Korea, Rep', 'Luxembourg', 'Mexico', 'Netherlands','New Zealand',
             'Norway', 'Poland', 'Portugal', 'Slovak Republic','Slovenia',
             'Spain', 'Sweden', 'Switzerland', 'Turkiye', 'United Kingdom','United States']
expedu = expedu[expedu['Country Name'].isin(countries)]
enrollment = enrollment[enrollment['Country Name'].isin(countries)]
literacy = literacy[literacy['Country Name'].isin(countries)]
meanschool = meanschool[meanschool['Country Name'].isin(countries)]
print(expedu.head())
print(enrollment.head())
print(literacy.head())
print(meanschool.head())
```

```python
drops = ['Country Code', 'Indicator Name', 'Indicator Code']
years = [str(year) for year in range(2000, 2025)]
years2 = [str(year) for year in range(2000, 2023)]
expedu = expedu.drop(columns=drops)
enrollment = enrollment.drop(columns=drops)
literacy = literacy.drop(columns=drops)
expedu = expedu.set_index('Country Name')
enrollment = enrollment.set_index('Country Name')
literacy = literacy.set_index('Country Name')
meanschool = meanschool.set_index('Country Name')
expedu = expedu[years]
enrollment = enrollment[years]
literacy = literacy[years]
meanschool = meanschool[years2]
print(expedu.head())
print(enrollment.head())
print(literacy.head())
print(meanschool.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))
for country in expedu.index:
    plt.plot(expedu.columns, expedu.loc[country], label=country)
plt.xlabel('')
plt.ylabel(' ')
plt.title('   ')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
for country in enrollment.index:
    plt.plot(enrollment.columns, enrollment.loc[country], label=country)
plt.xlabel('')
plt.ylabel(' ')
plt.title('   ')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
